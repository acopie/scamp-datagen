```python
def scale_data(data):
    """
    :return initial data, standard scaled data, minmax scaled data
    """
    _data = data.drop(columns=data.columns[0]) #drop column with problem name
    X = _data.to_numpy(dtype='float64')
    
    scalerStandard = StandardScaler().fit(X)
    X_std = scalerStandard.transform(X)

    scalerMinMax = MinMaxScaler().fit(X)
    X_MinMax = scalerMinMax.transform(X)
    return X, X_std, X_MinMax
```

```python
INPUT_PATH='../datasets/results/instances_characteristics/ianuarie2025'
OUTPUT_PATH='../datasets/results/ranks/runTime_1min_sa_ts_multilabel/'
OUTPUT_SUFIX='_1min_8alg'
```

```python
def read_data_for_embeding(data_representation, INPUT_PATH, add_statistics=False, label_alg = False):
    data, data_labels = read_info_graph(data_representation, INPUT_PATH)
    if add_statistics:
        data_statistics, data_statistic_labels = read_info_graph("statistics", INPUT_PATH)
        data = add_info_from_statistics_to_graph(data_statistics, data)

    if label_alg:
        label_data = read_output(dir_path=OUTPUT_PATH, sufix=OUTPUT_SUFIX)
        data_labels = build_algorithm_class_label(label_data)

    #remove instance name column
    data = data.drop(columns=[data.columns[0]])

    df = pd.DataFrame(data)
    
    #Drop columns with Null values
    cols = df.columns[df.isna().any()].tolist()
    if  cols:
        print("Columns with null values", cols)
    df = df.drop(columns=cols)
    
    df = df.to_numpy(dtype='float64')

    return df, data_labels
    
```

# TSNE

```python
def gen_embedding_TSNE(data_representation, n_components=2, perplexity=3, add_statistics=False, label_alg=False):
    data, data_labels = read_data_for_embeding(data_representation, INPUT_PATH, add_statistics, label_alg)
   
    X = data
    
    scalerStd = StandardScaler().fit(X)
    X_std = scalerStd.transform(X)

    scalerMinMax = MinMaxScaler().fit(X)
    X_minmax = scalerMinMax.transform(X)
    
    embeddingTSNE = TSNE(n_components=n_components, learning_rate='auto',init='pca', perplexity=perplexity)
    embeddingTSNE = embeddingTSNE.fit_transform(X)
    
    embeddingTSNE_std = TSNE(n_components=n_components, learning_rate='auto',init='pca', perplexity=perplexity)
    embeddingTSNE_std = embeddingTSNE_std.fit_transform(X_std)

    embeddingTSNE_minmax = TSNE(n_components=n_components, learning_rate='auto',init='pca', perplexity=perplexity)
    embeddingTSNE_minmax = embeddingTSNE_minmax.fit_transform(X_minmax)
    
    return data, embeddingTSNE, embeddingTSNE_std, embeddingTSNE_minmax, data_labels
```

```python
def TSNE_representation(type, title, add_statistics=False, label_alg=False, perplexity=3):
    data_graph, embeddingTSNE_graph, embeddingTSNE_graph_std, embeddingTSNE_graph_minmax, data_labels_graph = gen_embedding_TSNE(type, perplexity=perplexity, add_statistics=add_statistics,label_alg=label_alg)

    plt.style.use('ggplot')
    #plt.style.use('seaborn-v0_8-colorblind')

    fig, ax = plt.subplots(2)
    
    fig.suptitle(title + " std/minMax")

    data_names = data_labels_graph.drop_duplicates().reset_index(drop=True)
   
    # Factorize the column
    data_labels_graph['label_encoded'] = pd.factorize(data_labels_graph)[0]

    scatter = ax[0].scatter(embeddingTSNE_graph_std[:, 0], embeddingTSNE_graph_std[:, 1], c = data_labels_graph['label_encoded'])
    
    ax[0].set(xlabel='d1', ylabel='d2')

    ax[0].legend(scatter.legend_elements()[0], data_names,
          title="Problem",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
    
    scatter = ax[1].scatter(embeddingTSNE_graph_minmax[:, 0], embeddingTSNE_graph_minmax[:, 1], c = data_labels_graph['label_encoded'])
    
    ax[1].set(xlabel='d1', ylabel='d2')

    ax[1].legend(scatter.legend_elements()[0], data_names,
          title="Problem",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

    # for i, txt in enumerate(data.iloc[:,0].to_numpy()):
    #     ax.annotate(txt, (embeddingTSNE_MinMax[i, 0], embeddingTSNE_MinMax[i, 1]))
     
```
