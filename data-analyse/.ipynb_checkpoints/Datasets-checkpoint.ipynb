```python
import glob
from pathlib import Path
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
def values_output(input_directory, filter="*", columns=['graph_assortativity_quantity',
                            'graph_assortativity_s_execution_time',
                            'graph_assortativity_s_quantity',
                            'graph_assortativity_s_number'], interval=[-1,1]):
    """
    identifies problems where all allgorithms are equivalent
    """
    print(input_directory)
    ch_files = sorted(glob.glob(f'{input_directory}/{filter}'))

    for ch_file in ch_files:
        print(ch_file)
        data = pd.read_csv(ch_file)
        #histograma

        fig, axs = plt.subplots(2, 2, figsize=(10, 8))

        # Plot data on each subplot
        axs[0, 0].hist(data[columns[0]], range=interval)
        axs[0, 0].set_title (columns[0])
        axs[0, 1].hist(data[columns[1]], range=interval)
        axs[0, 1].set_title (columns[1])
        axs[1, 0].hist(data[columns[2]], range=interval)
        axs[1, 0].set_title (columns[2])
        axs[1, 1].hist(data[columns[3]], range=interval)
        axs[1, 1].set_title (columns[3])

        # Show the figure
        plt.show()

        plt.hist(data[columns], range=interval)
        plt.legend(columns)

       
```

```python
values_output('../datasets/results/instances_characteristics/ianuarie2025/operations', filter="*2asp*")
```
