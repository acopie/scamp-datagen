```python
def scale_data(data):
    """
    :return initial data, standard scaled data, minmax scaled data
    """
    _data = data.drop(columns=data.columns[0]) #drop column with problem name
    X = _data.to_numpy(dtype='float64')
    
    scalerStandard = StandardScaler().fit(X)
    X_std = scalerStandard.transform(X)

    scalerMinMax = MinMaxScaler().fit(X)
    X_MinMax = scalerMinMax.transform(X)
    return X, X_std, X_MinMax
```

# TSNE

```python
def gen_embedding_TSNE(data_representation, n_components=2, perplexity=3):
    data, data_labels = read_info_graph(data_representation, INPUT_PATH)

    _data = data.drop(columns=[data.columns[0]])
    #Drop columns with Null values
    df = pd.DataFrame(_data)
    cols = df.columns[df.isna().any()].tolist()
    if  cols:
        print("Columns with null values", cols)
    _data = _data.drop(columns=cols)
    
    X = _data.to_numpy(dtype='float64')
    
    scalerStd = StandardScaler().fit(X)
    X_std = scalerStd.transform(X)

    scalerMinMax = MinMaxScaler().fit(X)
    X_minmax = scalerMinMax.transform(X)
    
    embeddingTSNE = TSNE(n_components=n_components, learning_rate='auto',init='pca', perplexity=perplexity)
    embeddingTSNE = embeddingTSNE.fit_transform(X)
    
    embeddingTSNE_std = TSNE(n_components=n_components, learning_rate='auto',init='pca', perplexity=perplexity)
    embeddingTSNE_std = embeddingTSNE_std.fit_transform(X_std)

    embeddingTSNE_minmax = TSNE(n_components=n_components, learning_rate='auto',init='pca', perplexity=perplexity)
    embeddingTSNE_minmax = embeddingTSNE_minmax.fit_transform(X_minmax)
    
    return _data, embeddingTSNE, embeddingTSNE_std, embeddingTSNE_minmax, data_labels

def gen_embedding_TSNE_combined_with_statistics(data_representation, n_components=2, perplexity=3):
    data, data_labels = read_info_graph(data_representation, INPUT_PATH)
    data_statistics, data_statistic_labels = read_info_graph("statistics", INPUT_PATH)
    
    data = add_info_from_statistics_to_graph(data_statistics, data)
    _data = data.drop(columns=data.columns[0])

    df = pd.DataFrame(_data)
    cols = df.columns[df.isna().any()].tolist()
    if  cols:
        print("Columns with null values", cols)
    _data = _data.drop(columns=cols)

    X = _data.to_numpy()

    scalerStd = StandardScaler().fit(X)
    X_std = scalerStd.transform(X)

    scalerMinMax = MinMaxScaler().fit(X)
    X_minmax = scalerMinMax.transform(X)

    embeddingTSNE = TSNE(n_components=n_components, learning_rate='auto',init='pca', perplexity=perplexity)
    embeddingTSNE = embeddingTSNE.fit_transform(X)
    
    embeddingTSNE_std = TSNE(n_components=n_components, learning_rate='auto',init='pca', perplexity=perplexity)
    embeddingTSNE_std = embeddingTSNE_std.fit_transform(X_std)

    embeddingTSNE_minmax = TSNE(n_components=n_components, learning_rate='auto',init='pca', perplexity=perplexity)
    embeddingTSNE_minmax = embeddingTSNE_minmax.fit_transform(X_minmax)
    
    return _data, embeddingTSNE, embeddingTSNE_std, embeddingTSNE_minmax, data_labels
```

# UMAP

```python
def gen_embedding_UMAP(data_representation, n_components):
    data, data_labels = read_info_graph(data_representation, INPUT_PATH)
    _data = data.drop(columns=data.columns[0])

    df = pd.DataFrame(_data)
    cols = df.columns[df.isna().any()].tolist()
    if  cols:
        print("Columns with null values", cols)
    _data = _data.drop(columns=cols)
    
    X = _data.to_numpy()
    
    scalerStd = StandardScaler().fit(X)
    X_std = scalerStd.transform(X)

    scalerMinMax = MinMaxScaler().fit(X)
    X_minmax = scalerMinMax.transform(X)
    
    embeddingUMAP = umap.UMAP(n_neighbors=10, min_dist=0.1, n_components=n_components, metric='cosine',random_state=42).fit_transform(X)
    embeddingUMAP_std = umap.UMAP(n_neighbors=10, min_dist=0.1, n_components=n_components, metric='cosine',random_state=42).fit_transform(X_std)
    embeddingUMAP_minmax = umap.UMAP(n_neighbors=10, min_dist=0.1, n_components=n_components, metric='cosine',random_state=42).fit_transform(X_minmax)

    return _data, embeddingUMAP, embeddingUMAP_std, embeddingUMAP_minmax, data_labels

def gen_embedding_UMAP_stat(data_representation, n_components):
    data, data_labels = read_info_graph(data_representation, INPUT_PATH)
    data_statistics, data_statistic_labels = read_info_graph("statistics", INPUT_PATH)
    data = add_info_from_statistics_to_graph(data_statistics, data)
    
    _data = data.drop(columns=data.columns[0])

    df = pd.DataFrame(_data)
    cols = df.columns[df.isna().any()].tolist()
    if  cols:
        print("Columns with null values", cols)
    _data = _data.drop(columns=cols)
    
    X = _data.to_numpy()
    
    scalerStd = StandardScaler().fit(X)
    X_std = scalerStd.transform(X)

    scalerMinMax = MinMaxScaler().fit(X)
    X_minmax = scalerMinMax.transform(X)
    
    embeddingUMAP = umap.UMAP(n_neighbors=10, min_dist=0.1, n_components=n_components, metric='cosine',random_state=42).fit_transform(X)
    embeddingUMAP_std = umap.UMAP(n_neighbors=10, min_dist=0.1, n_components=n_components, metric='cosine',random_state=42).fit_transform(X_std)
    embeddingUMAP_minmax = umap.UMAP(n_neighbors=10, min_dist=0.1, n_components=n_components, metric='cosine',random_state=42).fit_transform(X_minmax)

    return _data, embeddingUMAP, embeddingUMAP_std, embeddingUMAP_minmax, data_labels
```
